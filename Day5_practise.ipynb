```python
# Synchronization - use for shared resource
from threading import Lock,Thread
import time
l=Lock()
def wish(name,age):
    l.acquire()
    print("Hi",name)
    time.sleep(2)
    print("Your age is",age)
    l.release()

t1=Thread(target=wish, args=("RAhul",15))
t2=Thread(target=wish, args=("Rushi",20))

t1.start()
t2.start()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
 
# Load the dataset
df = pd.read_csv('./DAY4/cars_clus.csv')
 
# Function to filter data by manufacturer
def filter_by_manufacturer(manufacturer):
    return df[df['manufact'] == manufacturer]
 
# Function to plot graphs
def plot_graphs(filtered_df):
    # Scatterplot: Engine size vs. Fuel efficiency
    plt.figure(figsize=(10, 5))
    plt.scatter(filtered_df['engine_s'], filtered_df['mpg'])
    plt.title('Engine Size vs. Fuel Efficiency')
    plt.xlabel('Engine Size (L)')
    plt.ylabel('Miles per Gallon (mpg)')
    plt.show()
 
    # Bar Chart: Average price by model
    avg_price = filtered_df.groupby('model')['price'].mean()
    plt.figure(figsize=(10, 5))
    avg_price.plot(kind='bar')
    plt.title('Average Price by Model')
    plt.xlabel('Model')
    plt.ylabel('Average Price ($)')
    plt.show()
 
    # Line Chart: Resale value vs. Sales
    plt.figure(figsize=(10, 5))
    plt.plot(filtered_df['sales'], filtered_df['resale'], marker='o')
    plt.title('Resale Value vs. Sales')
    plt.xlabel('Sales')
    plt.ylabel('Resale Value')
    plt.show()
 
    # Histogram: Distribution of horsepower
    plt.figure(figsize=(10, 5))
    plt.hist(filtered_df['horsepow'], bins=10, edgecolor='black')
    plt.title('Distribution of Horsepower')
    plt.xlabel('Horsepower')
    plt.ylabel('Frequency')
    plt.show()

    # Pie Chart : Market Share by Manufacturer
    sales_by_manufacturer=df.groupby('manufact')['sales'].sum().sort_values(ascending=False)
    top_n=5
    top_manufacturers=sales_by_manufacturer.head(top_n)
    other_sales=sales_by_manufacturer[top_n:].sum()
    top_manufacturers['others']=other_sales
    plt.figure(figsize=(8,8))
    plt.pie(top_manufacturers,labels=top_manufacturers.index,autopct='%1.1f%%')
    plt.title('Market Share by Manufacturer')
    plt.show()

   
 
# Example usage
selected_manufacturer = 'Toyota'  # This can be dynamically selected by the user
filtered_df = filter_by_manufacturer(selected_manufacturer)
plot_graphs(filtered_df)
```
