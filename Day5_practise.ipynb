```python

def calc_square(lst):
    print("Calculating Squares: ")
    for ele in lst:
        # cpu will be idle for 0.2 seconds
        time.sleep(0.2)
        print(f"Square of {ele}: {ele ** 2}")
       
 
 
def calc_cube(lst):
    print("Calculating Cubes: ")
    for ele in lst:
        time.sleep(0.2)
        print(f"Cube of {ele}: {ele ** 3}")
```

```python
# Synchronization - use for shared resource
from threading import Lock,Thread
import time
l=Lock()
def wish(name,age):
    l.acquire()
    print("Hi",name)
    time.sleep(2)
    print("Your age is",age)
    l.release()

t1=Thread(target=wish, args=("RAhul",15))
t2=Thread(target=wish, args=("Rushi",20))

t1.start()
t2.start()
```
