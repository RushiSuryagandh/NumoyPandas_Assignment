```python
#Question 2:How will you join the two data sets?
result=pd.merge(match,Deliveries,left_on="id",right_on="match_id")
result.columns
```

```python
#Question 4:What is the strike rate of BB McCullum in IPL 2008?
player = 'BB McCullum'  # The player's name
season = 2008  # The season of IPL

# Filter the data for BB McCullum in the 2008 season
mcCullum_data =result[(result['batsman'] == player) & (result['season'] == season)]

# Calculate total runs and total balls faced
total_runs = mcCullum_data['batsman_runs'].sum()
total_balls = mcCullum_data['ball'].count()  # count() gives the number of balls faced

# Calculate the strike rate
strike_rate = (total_runs / total_balls) * 100

print(f"BB McCullum's Strike Rate in IPL 2008: {strike_rate:.2f}")
```

```python
##Question 5:Which IPL Team concedes maximum extra runs?

# Group by the bowling team and sum total extra runs conceded
team_extras = Deliveries.groupby('bowling_team')['extra_runs'].sum().reset_index()
# Sort the result to find the team with the maximum extra runs conceded
team_extras_sorted = team_extras.sort_values(by='extra_runs', ascending=False)
team_extras_sorted.iloc[0]

```
